```python
sns.set_style("whitegrid")
plt.figure(figsize = (10,5))
spDataGroups = spData["race/ethnicity"].unique()
colors = ["blue", "orange", "green", "red", "purple"]


for color, group in zip(colors,spDataGroups):
    sns.distplot(spData[spData["race/ethnicity"]== group]["math score"])
    group_mean = spData[spData["race/ethnicity"]==group]["math score"].mean()
    plt.vlines(group_mean, 0, 0.035, colors = color)
    
plt.legend(spDataGroups)
```
